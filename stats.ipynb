```python
import pandas as pd
contrib = pd.read_csv("financial_contributions.csv")
profesori=pd.read_csv("profesori.csv")
contrib
contrib[['Date', 'Nume Prenume', 'Sursa', 'Currency', 'Incasat', 'NumPers','Clasa']]
```

```python
cost_per_pers=120+30
colegi = contrib.NumPers.sum()
profi = profesori["Vineri15"].sum()
flori= profesori["Flori"].sum()
flori_cost=flori*70
total_px=colegi+profi
incasat = contrib.Incasat.sum()
real_cost = total_px*cost_per_pers+flori_cost
spre_retur = incasat-real_cost

print(f'Persoane: {total_px}: fosti Studenti {colegi} si profesori {profi}')
print(f'RON incasat: {incasat}')
print(f'RON real_cost: {real_cost}')
print(f'RON flori_cost: {flori_cost} pentru {flori} buchete')
print(f'RON retur: {spre_retur}')
```
